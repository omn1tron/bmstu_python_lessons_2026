## Арифметические операции

#### Задание 1 (Чаевые в ресторане)

**На вход:** сумма счёта (целое число, например, 2350) и процент чаевых, который вы хотите оставить (целое число, например, 15).

**На выходе:** итоговая сумма к оплате (счёт + чаевые) и сумма чаевых в рублях.

Формат вывода: `Счёт: 2350 руб. Чаевые (15%): 352.5 руб. Итого: 2702.5 руб.`


#### Задание 2 (Площадь кольца)

**На вход:** два целых числа: внешний радиус (R) и внутренний радиус (r) кольца.

**На выходе:** строка с данными и результатом вычисления площади кольца (S = π*(R² - r²)).

Формат вывода: `Площадь кольца с радиусами 5 и 3 = 50.24`

````python
summ = int(input())
tip = int(input())
tip_s = summ * (tip/100)

print(f'Счёт: {summ} руб. Чаевые ({tip}%): {tip_s} руб. Итого: {summ + tip_s} руб.')
````

````python
R = int(input())
r = int(input())
S = 3.14 * (R**2 - r**2)

print(f'Площадь кольца с радиусами {R} и {r} = {S}')
````

#### Задание 1 (Генератор никнеймов)

Пользователь вводит имя и фамилию через пробел. Программа генерирует никнейм.

    Первые 3 буквы фамилии (в верхнем регистре).

    Символ подчеркивания.

    Первые 2 буквы имени (в нижнем регистре).

    Длина полного имени (количество символов, включая пробел).

**Пример:**

    Ввод: Марина Александрова
    
    Вывод: 'АЛЕ_ма10'
    
    Пояснение: АЛЕ (первые 3 от Александрова), ма (первые 2 от Марины в нижнем), 10 (в строке "Марина Александрова" 10 символов, считая пробел).

#### Задание 2 (Поиск хэштега)

**На вход:** строка текста, в которой спрятан хэштег (слово, начинающееся с символа #, например, #путешествия).

Гарантируется, что в строке есть ровно один хэштег и он отделён пробелами от других слов.

Формат ввода: `Смотрите мои фото из #отпуск на море`

**На выход:** вывести только этот хэштег, но без символа # и в верхнем регистре.

Формат вывода: `ОТПУСК`

````python
fio = input().split(sep=' ')
fa = fio[0]
pepe = fio[1]
watafa = len(fa + pepe)
print(f'{((fa[:3]).upper())}_{pepe[:2].lower()}{watafa}')
````

#### Задание 2 ("Угадай число")

Программа "загадывает" число от 1 до 100 (можно использовать переменную, например secret = 42).
Пользователю дается 5 попыток, чтобы угадать.

**На вход:** В цикле вводятся числа.

**На выходе:** После каждой попытки программа подсказывает "Больше" или "Меньше". Если число угадано — выводится "Поздравляю!" и цикл прерывается. Если попытки закончились — выводится "Попытки кончились. Было загадано: 42".

####  Задание 3 (Сумма цифр числа)

**На вход:** целое неотрицательное число.

**На выходе:** сумма его цифр.

**Условие:** Строки использовать нельзя.

**Пример:** 

    Ввод: 1234
    
    Вывод: 10

````python
secret = 52
for i in range(5):
    guess = int(input())
    if guess == secret:
        print("Поздравляю!")
        break

    if guess > secret:
        print("Больше")
    else:
        print("Меньше")

````

````python
i = int(input())
sum = 0
while i > 0:
    k = i % 10
    i = i //10
    sum += k
print(sum)
````
