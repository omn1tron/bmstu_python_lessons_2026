## Работа со списками

Дана матрица 3х2, пример:
```
matrix = [
    [1, 2],
    [5, 6],
    [9, 10],
]
```

Необходимо написать функцию, которая будет её транспонировать, т.е. приводить к виду
```
matrix = [
    [1, 5, 9],
    [2, 6, 10],
]
```

````python
matrix = [
    [1,2],
    [5,6],
    [9,10],
]

trans_mat = []
for i in range(2):
    r = [0] * 3
    trans_mat.append(r)

for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        trans_mat[j][i] = matrix[i][j]

print(trans_mat)
````

#### Задача "Рекомендации"

На вход:

список recom_ids = [1, 2, 3]

список seen_ids = [1, 4, 5]

На выход:

список filtered ?= [2, 3]

Условия:
1. надо составить новый список айдишников
2. в том же порядке что recom_ids
3. НЕ содержит seen_ids

Заготовка:
```python
from typing import List

def get_filtered(recom_ids: List[int], seen_ids: List[int]) -> List[int]:
    filtered = []
    # тут логика
    return filtered
```

Требуется оценить алгоритмическую сложность полученного решения

````python
from typing import List


recom_ids = [1, 2, 3]
seen_ids = [1, 4, 5]

def get_filtered(recom_ids: List[int], seen_ids: List[int]) -> List[int]:
    filtered = []
    for i in recom_ids:
        if i not in seen_ids:
            filtered.append(i)

    return filtered

print(get_filtered(recom_ids, seen_ids))
````
